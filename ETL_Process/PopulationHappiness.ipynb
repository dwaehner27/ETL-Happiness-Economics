```python
#dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Datasets/Happiness_by_country_2020.csv"
happiness_df = pd.read_csv(csv_file)
happiness_df.head()
```

```python
population = "Datasets/Population_by_Country_2020.csv"
population_df = pd.read_csv(population)
population_df.head()
```

```python
refined_happiness_df = happiness_df[["Country name", "Regional indicator", "Ladder score", "Logged GDP per capita", "Social support", "Perceptions of corruption"]].copy()
refined_happiness_df.head()
```

```python
refined_happiness_df.rename(columns={"Country name":"country_name", "Regional indicator":"region", "Ladder score":"score", "Logged GDP per capita":"gdp_per_capita", "Social support":"social_support", "Perceptions of corruption":"corruption_perception"},inplace=True)
```

```python
refined_happiness_df.reset_index(inplace=True)
```

```python
refined_happiness_df.rename(columns={"index":"id"},inplace=True)
```

```python
refined_happiness_df.head()
```

```python
refined_population_df = population_df [["Country (or dependency)", "Population (2020)", "Density (P/Km²)", "Land Area (Km²)", "Migrants (net)"]].copy()
refined_population_df.head()
```

```python
refined_population_df.rename(columns={"Country (or dependency)":"country_name", "Population (2020)":"population", "Density (P/Km²)":"density", "Land Area (Km²)":"land_area", "Migrants (net)":"migrants"},inplace=True)
```

```python
refined_population_df.reset_index(inplace=True)
```

```python
refined_population_df.rename(columns={'index':'id'},inplace=True)
```

```python
refined_population_df.head()
```

```python
happiness_dup =  refined_happiness_df['country_name'].value_counts().sort_values(ascending=False)
happiness_dup
```

```python
population_dup = refined_population_df["country_name"].value_counts().sort_values(ascending=False)
population_dup
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/happiness_vs_population"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
refined_happiness_df.to_sql(name='happiness',con=engine,if_exists='append',index=False)
```

```python
pd.read_sql_query('select * from happiness',con=engine).head()
```

```python
refined_population_df.to_sql(name='population',con=engine,if_exists='append',index=False)
```

```python
pd.read_sql_query('select * from population',con=engine).head()
```

```python
pd.read_sql_query('SELECT h.country_name, h.region, h.score, h.gdp_per_capita, h.social_support, h.corruption_perception, p.population, p.density, \
                    p.land_area, p.migrants FROM happiness h JOIN population p ON h.country_name = p.country_name ORDER BY h.score DESC',con=engine).head()
```
